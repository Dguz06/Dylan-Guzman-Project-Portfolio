```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.feature_extraction.text import *
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
from sklearn import metrics
```

```python
Gab=pd.read_csv("/content/GabHateCorpus_annotations.csv")
Gab=Gab.drop(["ID","Annotator"],axis=1) #cleaned up the unneccessary columns
Gab=Gab.dropna() #got rid of NA
```

```python
Gab[["REL","RAE","SXO","GEN","IDL","NAT","POL","MPH","EX", "IM"]]=Gab[["REL","RAE","SXO","GEN","IDL","NAT","POL","MPH","EX", "IM"]].astype(int)
```

```python
Gab.head()
```

```python
#SVM to classify what text is fits into each category

X = Gab["Text"]
y = Gab["HD"] #category trying to predict

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
cv    = CountVectorizer(max_features=1000, stop_words='english')
X_train = cv.fit_transform(X_train)

tfidf = TfidfTransformer(use_idf=True)
X_train = tfidf.fit_transform(X_train)

X_test = cv.transform(X_test)
X_test = tfidf.transform(X_test)

# Create a classifier model
classifier = SGDClassifier()

# Then train the model using the training sets
classifier = classifier.fit(X_train, y_train)

y_train.value_counts(normalize=True)
```

```python
y_pred_train = classifier.predict(X_train)
metrics.accuracy_score(y_train, y_pred_train)
```

```python
y_pred_test = classifier.predict(X_test)
metrics.accuracy_score(y_test, y_pred_test)
```

```python
print(metrics.classification_report(y_test, y_pred_test, zero_division=0))
```
